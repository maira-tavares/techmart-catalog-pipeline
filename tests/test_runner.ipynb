# NOTEBOOK: test_runner
- **Purpose:** Run all unit tests for Silver layer transformation functions inside Databricks.
- **Tests:** 37 unit tests covering weight normalization, price normalization and vendor name standardization.
- **Run:** On demand - after any change to transformation functions in `02_silver_standardize`.
- **Notes:** Test file is copied to `/tmp` before running because Databricks Repos filesystem does not allow pytest to create `__pycache__` directories.
- **Author:** Maira Tavares

```python
# Copy test file to /tmp and run from there
import subprocess
import shutil

# Copy test file to /tmp where Databricks allows file operations
shutil.copy(
    "/Workspace/Repos/ts.maira@gmail.com/techmart-catalog-pipeline/tests/test_transformations.py",
    "/tmp/test_transformations.py"
)

print("✅ Test file copied to /tmp")

# Run pytest from /tmp — no filesystem restrictions there
result = subprocess.run(
    [
        "python", "-m", "pytest",
        "/tmp/test_transformations.py",
        "-v"
    ],
    capture_output=True,
    text=True
)

print(result.stdout)

if result.stderr:
    print("STDERR:", result.stderr)

if result.returncode == 0:
    print("✅ ALL TESTS PASSED")
else:
    print("❌ SOME TESTS FAILED")
```
